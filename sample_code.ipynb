```python
#import necessary library
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
 
```

```python
data ={'hour_studied' : [1,2,3,4,5,6,7,8],
       'marks':[54,55,56,64,67,87,90,99]}
```

```python
df = pd.DataFrame(data)
```

```python
#splitting data into two input x and output y
x = df[['hour_studied']]
y = df['marks']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
#makinf prediction 
#add a new column for predicted marks
df['predicted']=model.predict(x)
```

```python
#plotting that data
plt.scatter(x,y,color = 'pink',label ='Actual marks')
#scatter of actual data 
plt.plot(x,df['predicted'],color ='black',label ='best fit line')
plt.xlabel("hours studied")
plt.ylabel("marks")
plt.legend()#shows actual marks and best fit lines 
plt.show()
```

```python
print("slope(m)",model.coef_[0])
print('intercept',model.intercept_)
print("r*2 score",model.score(x,y))
```
