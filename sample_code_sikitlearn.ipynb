```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#step 2 -create a sample adataset
data ={'hours_studied':[1,2,3,4,5,6,8,7,9,10],
'marks':[20,25,35,45,65,35,65,85,32,56]

}
```

```python
df = pd.DataFrame(data)
print("dataset ; \n",df)
```

```python
#step 3 - define input (x) and output (y)
#x-> independent cariable (hours studied ) kina vane hai le yo use garera marks determine garna lako xam

# y-> marks dependent variable kina vane hami le hours studied lai herera marks determine garxam
x = df[["hours_studied"]] #2d array kina vane yo feature ho
y = df['marks'] #1d array label vaera
```

```python
#step 4 - split for train and testing 
#train_test_split -> seperate the data into :training data and testing data
x_train ,x_test ,y_train,y_test=train_test_split(x,y,test_size= 0.2,random_state = 42)
""" x_train = training part of the input data 
x_test = testing part of the input data 
y_train = training part of the target variabel 
y_test = testing part of the target 
x,y why ? module jun import garexam tyo ay apply hunxa canera
test_size = 20%of the data is used for testing
random_state = control the randomness of spliting ,if you run your code 
then the it qill split data randomly .like 42 is standard numb
"""
```

```python
#step 5 - train the machine learning algorithm 
#linear regression -> best fitted line 
model = LinearRegression()
model.fit(x_train,y_train )#to train the model on training dataset
```

```python
#step 6 - testing this data 
#use test data (unseen by the model to check the prediction )
prediction = model.predict(x_test)
```

```python
print('\n test data : \n',x_test)
print('\n predict data : \n', prediction)
```

```python
#step 7 -predict marks for new student 
hours =[[11]]
predicted_marks = model.predict(hours)
print(f'\n if a student studies {hours[0][0]}',predicted_marks)
```
